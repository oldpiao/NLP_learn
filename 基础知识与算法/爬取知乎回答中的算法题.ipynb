```python
html_text = "./html/(6 封私信 _ 99+ 条消息) 算法到底应该怎么学？ - 知乎.html"
with open(html_text, 'r', encoding='utf-8') as f:
    html = f.read()
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "lxml")
a = soup.find_all('span', attrs={'class': 'RichText ztext CopyrightRichText-richText'})
```

```python
for n, i in enumerate(a[0].find_all('p')[32:52]):
    # print(n, [i.get_text()[:20]])
    print("- [%s](#%s)" % (i.find('b').get_text().replace(" ", "&nbsp;"), i.find('b').get_text().replace(" ", "&nbsp;")))
    for j in i.find_all('a'):
        print("\t- [%s](#%s)" % (j.get_text().replace(" ", "&nbsp;"), j.get_text().replace(" ", "&nbsp;")))
```

```python
for n, i in enumerate(a[0].find_all('p')[32:52]):
    # print(n, [i.get_text()[:20]])
    print("## %s" % i.find('b').get_text())
    for j in i.find_all('a'):
        print("### %s" % j.get_text().replace(" ", "&nbsp;"))
        print("[题目链接](%s)" % j['href'])
```

```python
import json

f_name = "算法系统学习-1.0 - 副本.ipynb"
f_name2 = "算法系统学习-1.1.ipynb"
with open(f_name, 'r', encoding='utf-8') as f:
    data = f.read()
    
data = json.loads(data)
for n, i in enumerate(data['cells']):
    if i['cell_type']=="markdown":
        for m, j in enumerate(i['source']):
            if j[:2] == "##":
                if m + 1 < len(data['cells'][n]['source']) and "题目链接" in data['cells'][n]['source'][m+1]:
                    data['cells'][n]['source'][m] += "\n[回到顶部](#目录)-"
                else:
                    data['cells'][n]['source'][m] += "\n[回到顶部](#目录)\n"
                
for i in data['cells']:
    if i['cell_type']=="markdown":
        for j in i['source']:
            if j[:2] == "##":
                print([j])
```

```python
with open(f_name2, 'w', encoding='utf-8') as f:
    f.write(json.dumps(data, ensure_ascii=False, indent=1))
```
